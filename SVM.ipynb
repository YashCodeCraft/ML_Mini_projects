# IRIS DATASET

```python


import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC #supprt vector classifier
```

```python


iris_info = load_iris()
iris_info
```

```python
data_iris = pd.DataFrame(iris_info['data'], columns=iris_info['feature_names'])
```

```python
data_iris['target'] = iris_info['target']
data_iris['Flower names'] = data_iris['target'].apply(lambda x : iris_info['target_names'] [x])
data_iris
```

```python
data_iris[data_iris['target']==1].head()
```

```python
data_iris0 = data_iris[:50]
data_iris1 = data_iris[50:100]
data_iris2 = data_iris[100:]
```

```python
data_iris1
```

```python
plt.scatter(data_iris0['sepal length (cm)'],data_iris0['sepal width (cm)'],marker='+',color='green')
plt.scatter(data_iris1['sepal length (cm)'],data_iris1['sepal width (cm)'],marker='+',color='blue')
```

```python
plt.scatter(data_iris0['petal length (cm)'],data_iris0['petal width (cm)'],marker='+',color='green')
plt.scatter(data_iris1['petal length (cm)'],data_iris1['petal width (cm)'],marker='+',color='blue')
plt.scatter(data_iris2['petal length (cm)'],data_iris2['petal width (cm)'],marker='+',color='red')
```

```python
# we should consider petal graph because of the near data points around linear line(High gamma with more accuracy)
# we should consider petal as a input
```

```python
input_=data_iris.drop(columns=['sepal width (cm)','sepal length (cm)','Flower names','target'])
output_=data_iris['target']
```

```python
input_train, input_test, output_train, output_test = train_test_split(input_, output_,train_size=0.8)
```

```python
print(len(input_train))
print(len(input_test))
```

```python
model = SVC(gamma=1)
model
```

```python
model.fit(input_train,output_train)
model
```

```python
model.score(input_test,output_test)
```

```python
model.predict([[4.7,1.5]])
```

```python
model1 = SVC(C=1)
model1.fit(input_train,output_train)
model1.score(input_test,output_test)
```
